```python
# Tuples 
# Tuples are an ordered sequence.
# Tuples are immutable, meaning they cannot be changed.
# Tuples are defined using parentheses and comma-seperated elements.
# Tuples are faster than lists.
# Types of tuples: str, int, float

# An example of a tuple
ratings = (10, 9, 6, 5, 10, 8, 9, 6, 2) 
tuple1 = ("disco", 10, 1.2) # In python, we can use negative index. Like, "disco"[-3], 10 [-2], 1.2 [-1]

# we can add 
tuple2 = tuple1 + ("hard rock", 10)
print(tuple2)

# we can slice
tuple3 = tuple2[0:3] # first three elements
print(tuple3)

# sorted
RatingsSorted = sorted(ratings)
print(RatingsSorted)

# Nested Tuple
NT = (1,2,("pop", "rock"), (3,4),("disco",(1,2)))
print(NT[2])   # ('pop', 'rock')
print(NT[2][1]) # rock

# Length of tuple
print(len(NT)) # 5

# Concatenate tuples
NT1 = NT + (5,6)
print(NT1)

# Repetition
NT2 = NT * 2
print(NT2)

# Membership
print(3 in NT) # False
```

```python
# Lists
# Lists are mutable, meaning they can be changed.
# Lists are defined using square brackets and comma-seperated elements.
# Lists are slower than tuples.

# An example of a list
#       0               1      2
L = ["Michael Jackson", 10.1, 1982]
print(L)
print(L[0:2])

L.extend(["pop", 10]) # Add elements to the list
print(L)

# Or We can concatenate
L1 = L + ["pop", 10] # Concatenate lists
print(L1)

# Change the element based on index
L[0] = "Hard Rock"
print(L)

# Delete the element based on index
del(L[0])
print(L)

# Split the string (separa a string em uma lista)
"hard rock".split() # ['hard', 'rock'] - default is space
"Ateez,Boa,CL,Dahyun".split(",") # ['hard', 'rock'] - split by comma

# Clone the list
L2 = L[:]
print(L2)

help(L2)
```

```python
# Sets
# Sets are a type of collection
   # This means that like lists and tuples you can input different Python types
# Unlike lists and tuples they are unordered
   # This means sets do not record element position
# Sets only have unique elements
   # This means there is only one of a particular element in a set

# An example of a set
Set1 = {"pop", "rock", "soul", "hard rock", "rock", "R&B", "rock", "disco"}
print(Set1)

# Convert list to set
album_list = ["Michael Jackson", "Thriller", 1982, "00:42:19", "pop, rock, R&B", 46.0, 65, "30-Nov-82", None, 10.0]
album_set = set(album_list)
print(album_set)

# Add element to set
kpop_group_random = {"BTS", "EXO", "ATEEZ", "GOT7"}
print(kpop_group_random)
kpop_group_random.add("TWICE")
kpop_group_random.add("G(I)-DLE")
print(kpop_group_random)
kpop_group_random.remove("GOT7")
print(kpop_group_random)

# Verify if the element is in the set
print("BTS" in kpop_group_random) # True
print("BLACKPINK" in kpop_group_random) # False

# Intersection
kpop_girl_group = {"TWICE", "BLACKPINK", "G(I)-DLE", "ITZY", "NEW JEANS"}
print(kpop_girl_group)

kpop_group = kpop_group_random & kpop_girl_group
print(f"Intersection: {kpop_group}")

# issubset - Check if the set is a subset of another set
print(kpop_group.issubset(kpop_group_random)) # True
```

```python
S = {"A", "B", "C"}
U = {"A", "Z", "C"}
print(S.union(U)) # {'A', 'B', 'C', 'Z'}
```

```python
# Dictionaries
# Dictionaries are a type of collection in Python
# This means that like lists and tuples you can input different Python types
# Unlike lists and tuples they are unordered
# This means dictionaries do not record element position
# Dictionaries have keys and values
# It is best to think of a dictionary as an unordered set of key: value pairs, with the requirement that the keys are unique (within one dictionary)
# A pair of braces creates an empty dictionary: {}
# Placing a comma-separated list of key:value pairs within the braces adds initial key:value pairs to the dictionary; this is also the way dictionaries are written on output

# An example of a dictionary
kpop_dict_leaders = {"BTS": "RM", "EXO": "Suho", "ATEEZ": "Hongjoong", "GOT7": "JB"}
print(kpop_dict_leaders["ATEEZ"])

# Add a new key-value pair
kpop_dict_leaders["TWICE"] = "Jihyo"
print(kpop_dict_leaders)

# Delete a key-value pair
del(kpop_dict_leaders["GOT7"])
print(kpop_dict_leaders)

# Verify if the key is in the dictionary
print("TWICE" in kpop_dict_leaders) # True
print("BLACKPINK" in kpop_dict_leaders) # False

D = {'a':0,'b':1,'c':2}

print(D.values())
```
